VISUALIZATION FREQUENCY DISTRIBUTION

```python
from nltk.tokenize import word_tokenize
```

```python
file=open('opinion.txt','r')
text=file.read()
```

```python
words = word_tokenize(text)
len(words)
```

```python
from nltk.probability import FreqDist

all_fdist=FreqDist(words).most_common(20)
print(all_fdist)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

all_fdist=pd.Series(dict(all_fdist))
fig,ax=plt.subplots(figsize=(6,6))

all_fdist.plot(kind='bar')
plt.title("frquency distribution of words")
plt.ylabel("count")
```

lowercase

```python
text =text.lower()  
```

removing special character

```python
import re
text = re.sub('[^A-Za-z0-9]+', ' ', text)
```

removiung words with numbers from data

```python
text = re.sub("\S*\d\S*", "", text).strip()
```

```python
import nltk
nltk.download('stopwords')
```

```python
import nltk
words = word_tokenize(text)
stopwords = nltk.corpus.stopwords.words('english')
words_sw_removed = []

for word in words:
    if word in stopwords:
        pass
    else:
        words_sw_removed.append(word)
```

```python
all_fdist = FreqDist(words_sw_removed).most_common(20)

all_fdist = pd.Series(dict(all_fdist))

fig, ax = plt.subplots(figsize=(7,7))

all_fdist.plot(kind='bar')
plt.title("Frequency Distribution of words")
plt.ylabel("Count")
```
